```python
import requests
import json
from pandas.io.json import json_normalize
```

```python
url = 'https://api.meraki.com/api/v0/networks/N_638948197133241698'
headers = {
  'Accept': '*/*',
    "x-cisco-meraki-api-key" : "b625da6fedf3a8092f4cde15973c9a3c0ca74000"
}
response = requests.request('GET', url, headers = headers)
print(response.text)
```

```python
data=json.loads(response.text)
```

```python
frame=json_normalize(data)
```

```python
frame
```
